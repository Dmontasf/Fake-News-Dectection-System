```python
data_merge = pd.concat([fake,true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
def processed_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]',"",text)
    text = re.sub("\\W"," ",text)
    text = re.sub("https?:://\S+|www\.\S+","",text)
    text = re.sub("<.*?>+","",text)
    text = re.sub("[%s]" % re.escape(string.punctuation),"",text)
    text = re.sub("\w*\d\w*","",text)
    return text
```

```python
data['text'] = data['text'].apply(processed_text)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(xv_train, y_train)
```

```python
pred_lr = lr.predict(xv_test)
```

```python
lr.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(xv_train, y_train)
```

```python
pred_dt = dt.predict(xv_test)
```

```python
dt.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state = 0)
gb.fit(xv_train, y_train)
```

```python
pred_gb = gb.predict(xv_test)
```

```python
gb.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 0)
rf.fit(xv_train, y_train)
```

```python
pred_rf = rf.predict(xv_test)
```

```python
rf.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output(n):
    if n == 0:
        return "This Article is fake"
    elif n == 1:
        return "This Article is NOT fake"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test['text'].apply(processed_text)
    new_x_test = new_def_test['text']
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = lr.predict(new_xv_test)
    pred_dt = dt.predict(new_xv_test)
    pred_gb = gb.predict(new_xv_test)
    pred_rf = rf.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(output(pred_lr[0]),
                                                                                                            output(pred_dt[0]),
                                                                                                            output(pred_gb[0]),
                                                                                                            output(pred_rf[0])))
    
```
